```python
X = df[['RM','LSTAT','PTRATIO']]           #features
y = df['MEDV']                             #our targeted variable
```

# Splitting the Data into Training & Testing 

```python
from sklearn.model_selection import train_test_split
```

cheat note: if we type 'train_test_split' and press shift+tab, then we can see how many variables we can import in this function.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

# Training Algorithm

```python
from sklearn.linear_model import LinearRegression  
```

```python
regressor = LinearRegression()                      #creating a varible and storing object in it that is LinearRegression
```

```python
regressor.fit(X_train, y_train)                     #We fit the model into training
```

# Predicting the Model

```python
predictions = regressor.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
```

```python
print(X_test)
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})         #comparing between test and predicted data
df
```

# Evaluation of Model

### Accuracy of the Model

```python
from sklearn import metrics
metrics.r2_score(y_test, y_pred)
```

# Error Measure

```python
print("Mean Square Error :", metrics.mean_squared_error(y_test,y_pred))
```

```python
print("Root Mean Square Error :", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
print("Mean Absolute Error :", metrics.mean_absolute_error(y_test, y_pred))
```
